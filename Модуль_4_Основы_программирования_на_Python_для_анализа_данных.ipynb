В переменной base_price хранится базовая цена товара, а в переменной growth_percent - процент увеличения цены (например, 20
означает рост на 20%).

Вычислите новую цену с учётом роста и сохраните её в переменную new_price.

Также вычислите размер прибавки (разницу между новой и старой ценой) и сохраните её в переменную delta.

Входные данные:

base_price = 100

growth_percent = 20

Выходные данные: Программа должна вычислить:
новую цену new_price
разницу между ценами delta

new_price = 120.0

delta = 20.0
